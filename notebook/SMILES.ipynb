# SMILES

This notebook is example for **SMILES-related** chemoinfomatic python modules:

- `rdkit`: SMILES -> Mol -> Draw
- `pubchempy` : NAME -> SMILES

## SMILES -> Mol -> Draw

````python
from rdkit import Chem
from rdkit.Chem import Draw
````

````python
SMILES_benzene = "C1=CC=CC=C1" # SMILES Object.
Mol_benzene = Chem.MolFromSmiles(SMILES_benzene) # Mol Object.
Draw.MolToImage(mol=Mol_benzene, size=(300, 300))
````

````python
Draw.MolsToGridImage(mols=[Mol_benzene])
````

## Compound name -> SMILES

````python
import pubchempy as pcp
````

````python
# Retrieve the specified compound records from PubChem.
compound_records = pcp.get_compounds(identifier='benzene', namespace='name')
SMILES_benzene = compound_records[0].isomeric_smiles # SMILES Object.
Mol_benzene = Chem.MolFromSmiles(SMILES_benzene) # Mol Object.
Draw.MolsToGridImage(mols=[Mol_benzene])
````

````python
SMILES_anthracene = pcp.get_compounds('anthracene', 'name')[0].isomeric_smiles
Mol_anthracene = Chem.MolFromSmiles(SMILES_anthracene)
Draw.MolsToGridImage(mols=[Mol_anthracene])
````

````python
Draw.MolsToGridImage(mols=[Mol_benzene, Mol_anthracene])
````
